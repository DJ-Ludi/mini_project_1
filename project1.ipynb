## Mini Project_1 (할인율 모니터링)

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
ch_driver = webdriver.Chrome('c://Python37/chromedriver.exe')
```

```python
lotte_list_url = "http://www.lotte.com/search/searchMain.lotte?tq=&PAGECOUNT=84&init=Y&requery=%EB%B9%88%ED%8F%B4%EB%A0%88%EC%9D%B4%EB%94%94%EC%8A%A4&MULTISHOPCATES=LVL1_CATE_NO&SRCHFIELD=&tracking=MH_SEARCH&imgSearchVal=%2Fplanshop%2FviewPlanShopDetail.lotte%3Fspdp_no%3D5372527%26tracking%3DMH_SEARCH_01_PLANSHOP&dpml_no=1&site_no=1&query=%EB%B9%88%ED%8F%B4%EB%A0%88%EC%9D%B4%EB%94%94%EC%8A%A4"
ch_driver.get(lotte_list_url)
```

```python
lotte_item_xpath = '//*[@id="searchForm2"]/div/div[2]/div[4]/div[1]/div/ul/li'
```

```python
#get items in page
results = ch_driver.find_elements_by_xpath(lotte_item_xpath)
```

```python
len(results)
```

```python
#Function for scraping title & discount rate

def get_sale_list():
    result_list = []
    results = ch_driver.find_elements_by_xpath(lotte_item_xpath)
    for item in results:
        try:
            subject = item.find_element_by_class_name('contents').text
            dis_rate = item.find_element_by_class_name('blindSale').text.strip(' %') 
            print(subject, sale_rate, sep='\n')
            result_list.append((
                subject,
                dis_rate
            ))
        except:
            pass
    return result_list
```

```python
#Scraping Next page

from time import sleep


ch_driver.implicitly_wait(5) # 요소가 로드되지 않았을 경우, 5초만큼 기다림

all_result = []
for i in range(1,5):
    item_data = get_sale_list()
    all_result.append(item_data)
    page_btn = ch_driver.find_element_by_xpath('//*[@id="searchForm2"]/div/div[2]/div[4]/div[2]/a[{page_num}]'.format(page_num = i))
    page_btn.click()
    sleep(1)
```

```python
all_result
```

```python
import re
```

```python
m = re.compile(r"[B][F][8][a-zA-Z0-9_]+")    #품번조건정의
product_list = []
for i in range(0,len(all_result)):
    for j in range(0,len(all_result[i])):
        try:
            a = m.search(all_result[i][j][0]).group()      #스트링으로 저장하고싶은데 Group() 으로 하면 에러남 -> 에러핸들링으로 해결 
            b = int(all_result[i][j][1])
            c = (a,b)
            product_list.append(c)
        except:
            pass
```

```python
product_list
```

```python
len(product_list)
```

```python
#Count violation by channels & List up 

max_rate = 20
mall = 0
mall_list = []
shop = 0
shop_list = []
for i in range(0,len(product_list)-1):
    if product_list[i][1] > max_rate :
        if len(product_list[i][0]) == 10:
            mall += 1
            mall_list.append(product_list[i])
        else:
            shop += 1
            shop_list.append(product_list[i])
    else:
        pass
print('mall = {}건,shop = {}건'.format(mall,shop))
print('몰 품번리스트 = ', mall_list)
print('매장 품번리스트 =', shop_list)
```
